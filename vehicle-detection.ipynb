```python
import os
import cv2
import numpy as np
import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.utils.data import Dataset, DataLoader
from torchvision import transforms
from pathlib import Path
import matplotlib.pyplot as plt
import matplotlib.patches as patches

# ── Classes we care about ────────────────────────────────
# From sakshamjn: auto=0, bus=1, car=2, lcv=3, 
#                 motorcycle=4, multiaxle=5, tractor=6, truck=7
# We keep: car(2), motorcycle(4), bus(1), truck(7)
# Remap to: car=0, motorcycle=1, bus=2, truck=3

KEEP_CLASSES = {2: 0, 4: 1, 1: 2, 7: 3}
CLASS_NAMES  = ['car', 'motorcycle', 'bus', 'truck']
NUM_CLASSES  = 4

IMG_SIZE = 416   # standard detection input size
device   = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f"Device: {device}")
print(f"Classes: {CLASS_NAMES}")

# ── Dataset class ─────────────────────────────────────────
class VehicleDataset(Dataset):
    def __init__(self, img_dir, label_dir, img_size=416, augment=False):
        self.img_size  = img_size
        self.augment   = augment
        self.samples   = []

        # Match images to labels
        img_files = {Path(f).stem: os.path.join(img_dir, f)
                     for f in os.listdir(img_dir)
                     if f.lower().endswith(('.jpg','.png','.jpeg'))}

        label_files = {Path(f).stem: os.path.join(label_dir, f)
                       for f in os.listdir(label_dir)
                       if f.endswith('.txt') and f != 'classes.txt'}

        for stem in label_files:
            if stem in img_files:
                self.samples.append((img_files[stem], label_files[stem]))

        print(f"  Found {len(self.samples)} image-label pairs")

    def __len__(self):
        return len(self.samples)

    def load_labels(self, label_path):
        boxes = []
        with open(label_path) as f:
            for line in f:
                parts = line.strip().split()
                if len(parts) != 5:
                    continue
                cls = int(float(parts[0]))
                if cls not in KEEP_CLASSES:
                    continue
                new_cls = KEEP_CLASSES[cls]
                x, y, w, h = map(float, parts[1:])
                boxes.append([new_cls, x, y, w, h])
        return boxes

    def __getitem__(self, idx):
        img_path, label_path = self.samples[idx]

        # Load image
        img = cv2.imread(img_path)
        if img is None:
            img = np.zeros((self.img_size, self.img_size, 3), dtype=np.uint8)
        img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
        img = cv2.resize(img, (self.img_size, self.img_size))

        # Augment
        if self.augment and np.random.rand() > 0.5:
            img = np.fliplr(img).copy()

        # Normalize
        img = img.astype(np.float32) / 255.0
        img = torch.FloatTensor(img).permute(2, 0, 1)  # (3, H, W)

        # Load labels
        boxes = self.load_labels(label_path)

        # Pad to fixed size for batching (max 20 boxes per image)
        MAX_BOXES = 20
        target = torch.zeros(MAX_BOXES, 5)  # (class, x, y, w, h)
        if boxes:
            boxes = torch.FloatTensor(boxes[:MAX_BOXES])
            target[:len(boxes)] = boxes

        return img, target

# ── Load datasets ─────────────────────────────────────────
print("\nLoading sakshamjn dataset:")
sak_base  = '/kaggle/input/datasets/sakshamjn/vehicle-detection-8-classes-object-detection/train'
sak_imgs  = os.path.join(sak_base, 'images') if os.path.exists(
             os.path.join(sak_base, 'images')) else sak_base
sak_lbls  = os.path.join(sak_base, 'labels')

# Check if images folder exists
print(f"  Images path: {sak_imgs}")
print(f"  Labels path: {sak_lbls}")
print(f"  Images exist: {os.path.exists(sak_imgs)}")

# List what's in train folder
print(f"  Train contents: {os.listdir(sak_base)}")
```

```python
import torch
import torch.nn as nn

# ── Building blocks ───────────────────────────────────────
class ConvBNReLU(nn.Module):
    def __init__(self, in_ch, out_ch, k=3, s=1, p=1):
        super().__init__()
        self.block = nn.Sequential(
            nn.Conv2d(in_ch, out_ch, k, s, p, bias=False),
            nn.BatchNorm2d(out_ch),
            nn.LeakyReLU(0.1, inplace=True)
        )
    def forward(self, x):
        return self.block(x)

class ResBlock(nn.Module):
    def __init__(self, ch):
        super().__init__()
        self.block = nn.Sequential(
            ConvBNReLU(ch, ch//2, k=1, p=0),
            ConvBNReLU(ch//2, ch, k=3, p=1)
        )
    def forward(self, x):
        return x + self.block(x)

# ── Backbone ──────────────────────────────────────────────
class Backbone(nn.Module):
    def __init__(self):
        super().__init__()
        self.stage1 = nn.Sequential(
            ConvBNReLU(3,   32,  k=3, s=1),
            ConvBNReLU(32,  64,  k=3, s=2),
            ResBlock(64),
        )
        self.stage2 = nn.Sequential(
            ConvBNReLU(64,  128, k=3, s=2),
            ResBlock(128), ResBlock(128),
        )
        self.stage3 = nn.Sequential(
            ConvBNReLU(128, 256, k=3, s=2),
            ResBlock(256), ResBlock(256), ResBlock(256),
        )
        self.stage4 = nn.Sequential(
            ConvBNReLU(256, 512, k=3, s=2),
            ResBlock(512), ResBlock(512),
        )
        self.stage5 = nn.Sequential(
            ConvBNReLU(512, 1024, k=3, s=2),
            ResBlock(1024),
        )

    def forward(self, x):
        s1 = self.stage1(x)
        s2 = self.stage2(s1)
        s3 = self.stage3(s2)
        s4 = self.stage4(s3)
        s5 = self.stage5(s4)
        return s3, s4, s5

# ── Detection Head ────────────────────────────────────────
class DetectionHead(nn.Module):
    def __init__(self, num_classes=4, num_anchors=3):
        super().__init__()
        self.num_classes = num_classes
        self.num_anchors = num_anchors
        out_ch = num_anchors * (5 + num_classes)  # 27

        # Large scale (13x13)
        self.head_large = nn.Sequential(
            ConvBNReLU(1024, 512,  k=1, p=0),
            ConvBNReLU(512,  1024, k=3),
            ConvBNReLU(1024, 512,  k=1, p=0),
            ConvBNReLU(512,  1024, k=3),
            ConvBNReLU(1024, 512,  k=1, p=0),
        )
        self.out_large  = nn.Conv2d(512, out_ch, 1, 1, 0)

        # Medium scale (26x26)
        self.lateral1   = ConvBNReLU(1024, 256, k=1, p=0)
        self.upsample1  = nn.Upsample(scale_factor=2)
        self.head_med   = nn.Sequential(
            ConvBNReLU(256+512, 256, k=1, p=0),
            ConvBNReLU(256,     512, k=3),
            ConvBNReLU(512,     256, k=1, p=0),
        )
        self.out_med    = nn.Conv2d(256, out_ch, 1, 1, 0)

        # Small scale (52x52)
        self.lateral2   = ConvBNReLU(256, 128, k=1, p=0)
        self.upsample2  = nn.Upsample(scale_factor=2)
        self.head_small = nn.Sequential(
            ConvBNReLU(128+256, 128, k=1, p=0),
            ConvBNReLU(128,     256, k=3),
            ConvBNReLU(256,     128, k=1, p=0),
        )
        self.out_small  = nn.Conv2d(128, out_ch, 1, 1, 0)

    def forward(self, s3, s4, s5):
        # Large
        feat_large = self.head_large(s5)
        out_large  = self.out_large(feat_large)

        # Medium
        lat1       = self.lateral1(s5)
        up1        = self.upsample1(lat1)
        med_feat   = torch.cat([up1, s4], dim=1)
        med_feat   = self.head_med(med_feat)
        out_med    = self.out_med(med_feat)

        # Small
        lat2       = self.lateral2(med_feat)
        up2        = self.upsample2(lat2)
        small_feat = torch.cat([up2, s3], dim=1)
        small_feat = self.head_small(small_feat)
        out_small  = self.out_small(small_feat)

        return out_large, out_med, out_small

# ── Full Model ────────────────────────────────────────────
class VehicleDetector(nn.Module):
    def __init__(self, num_classes=NUM_CLASSES):
        super().__init__()
        self.backbone = Backbone()
        self.head     = DetectionHead(num_classes)

    def forward(self, x):
        s3, s4, s5        = self.backbone(x)
        large, med, small = self.head(s3, s4, s5)
        return large, med, small

# Build and test
model_det    = VehicleDetector().to(device)
total_params = sum(p.numel() for p in model_det.parameters())
print(f"✅ VehicleDetector built from scratch")
print(f"   Total parameters: {total_params:,}")

dummy = torch.zeros(1, 3, IMG_SIZE, IMG_SIZE).to(device)
with torch.no_grad():
    out_l, out_m, out_s = model_det(dummy)
print(f"\nOutput shapes:")
print(f"  Large (13x13) : {out_l.shape}")
print(f"  Medium(26x26) : {out_m.shape}")
print(f"  Small (52x52) : {out_s.shape}")
```

```python
# ── Anchors (predefined box sizes) ───────────────────────
# Tuned for vehicle detection at 416x416
ANCHORS = {
    'large' : [(116,90), (156,198), (373,326)],  # large vehicles
    'medium': [(30,61),  (62,45),   (59,119)],   # medium
    'small' : [(10,13),  (16,30),   (33,23)],    # small/far vehicles
}

def build_targets(predictions, targets, anchors, grid_size, img_size=416):
    """
    Convert ground truth boxes to grid-relative targets
    targets: (B, MAX_BOXES, 5) — (class, cx, cy, w, h) normalized 0-1
    """
    B        = predictions.shape[0]
    n_anchors= len(anchors)
    out_ch   = 5 + NUM_CLASSES

    pred     = predictions.view(B, n_anchors, out_ch,
                                grid_size, grid_size)
    pred     = pred.permute(0,1,3,4,2)  # (B, A, G, G, 5+C)

    # Targets
    obj_mask  = torch.zeros(B, n_anchors, grid_size, grid_size,
                            dtype=torch.bool).to(device)
    noobj_mask= torch.ones( B, n_anchors, grid_size, grid_size,
                            dtype=torch.bool).to(device)
    tx = torch.zeros(B, n_anchors, grid_size, grid_size).to(device)
    ty = torch.zeros(B, n_anchors, grid_size, grid_size).to(device)
    tw = torch.zeros(B, n_anchors, grid_size, grid_size).to(device)
    th = torch.zeros(B, n_anchors, grid_size, grid_size).to(device)
    tcls= torch.zeros(B, n_anchors, grid_size, grid_size,
                      NUM_CLASSES).to(device)

    anch_t = torch.FloatTensor(anchors).to(device)

    for b in range(B):
        for t in range(targets.shape[1]):
            if targets[b, t, 3] == 0:
                continue   # no box

            cls   = int(targets[b, t, 0])
            gx    = targets[b, t, 1] * grid_size
            gy    = targets[b, t, 2] * grid_size
            gw    = targets[b, t, 3] * img_size
            gh    = targets[b, t, 4] * img_size

            gi, gj = int(gx), int(gy)
            gi = min(gi, grid_size-1)
            gj = min(gj, grid_size-1)

            # Find best anchor
            gt_box  = torch.FloatTensor([0,0,gw,gh]).to(device)
            anch_boxes = torch.cat([torch.zeros(n_anchors,2).to(device),
                                    anch_t], dim=1)
            iou_anch = bbox_iou(gt_box.unsqueeze(0),
                                anch_boxes, x1y1x2y2=False)
            best_a   = iou_anch.argmax()

            obj_mask  [b, best_a, gj, gi] = True
            noobj_mask[b, best_a, gj, gi] = False

            tx[b, best_a, gj, gi] = gx - gi
            ty[b, best_a, gj, gi] = gy - gj
            tw[b, best_a, gj, gi] = torch.log(
                gw / anch_t[best_a, 0] + 1e-16)
            th[b, best_a, gj, gi] = torch.log(
                gh / anch_t[best_a, 1] + 1e-16)
            tcls[b, best_a, gj, gi, cls] = 1

    return pred, obj_mask, noobj_mask, tx, ty, tw, th, tcls

def bbox_iou(box1, box2, x1y1x2y2=True):
    if not x1y1x2y2:
        b1_x1 = box1[:,0] - box1[:,2]/2
        b1_x2 = box1[:,0] + box1[:,2]/2
        b1_y1 = box1[:,1] - box1[:,3]/2
        b1_y2 = box1[:,1] + box1[:,3]/2
        b2_x1 = box2[:,0] - box2[:,2]/2
        b2_x2 = box2[:,0] + box2[:,2]/2
        b2_y1 = box2[:,1] - box2[:,3]/2
        b2_y2 = box2[:,1] + box2[:,3]/2
    else:
        b1_x1,b1_y1,b1_x2,b1_y2 = box1[:,0],box1[:,1],box1[:,2],box1[:,3]
        b2_x1,b2_y1,b2_x2,b2_y2 = box2[:,0],box2[:,1],box2[:,2],box2[:,3]

    inter_x1 = torch.max(b1_x1, b2_x1)
    inter_y1 = torch.max(b1_y1, b2_y1)
    inter_x2 = torch.min(b1_x2, b2_x2)
    inter_y2 = torch.min(b1_y2, b2_y2)
    inter    = torch.clamp(inter_x2-inter_x1, 0) * \
               torch.clamp(inter_y2-inter_y1, 0)

    b1_area  = (b1_x2-b1_x1) * (b1_y2-b1_y1)
    b2_area  = (b2_x2-b2_x1) * (b2_y2-b2_y1)
    union    = b1_area + b2_area - inter + 1e-16
    return inter / union

def compute_loss(predictions, targets, anchors, grid_size):
    pred, obj_mask, noobj_mask, tx, ty, tw, th, tcls = \
        build_targets(predictions, targets, anchors, grid_size)

    B, A, G, _, out_ch = pred.shape

    pred_x   = torch.sigmoid(pred[..., 0])
    pred_y   = torch.sigmoid(pred[..., 1])
    pred_w   = pred[..., 2]
    pred_h   = pred[..., 3]
    pred_obj = torch.sigmoid(pred[..., 4])
    pred_cls = pred[..., 5:]

    bce  = nn.BCELoss()
    mse  = nn.MSELoss()

    loss_x   = mse(pred_x[obj_mask],   tx[obj_mask])
    loss_y   = mse(pred_y[obj_mask],   ty[obj_mask])
    loss_w   = mse(pred_w[obj_mask],   tw[obj_mask])
    loss_h   = mse(pred_h[obj_mask],   th[obj_mask])
    loss_obj = bce(pred_obj[obj_mask],
                   torch.ones_like(pred_obj[obj_mask]))
    loss_noobj = bce(pred_obj[noobj_mask],
                     torch.zeros_like(pred_obj[noobj_mask]))
    loss_cls = bce(torch.sigmoid(pred_cls[obj_mask]),
                   tcls[obj_mask])

    total = (loss_x + loss_y + loss_w + loss_h +
             loss_obj + 0.5*loss_noobj + loss_cls)
    return total, loss_x, loss_y, loss_obj, loss_cls

# ── Training loop ─────────────────────────────────────────
optimizer_det = torch.optim.Adam(model_det.parameters(),
                                  lr=1e-3, weight_decay=1e-4)
scheduler_det = torch.optim.lr_scheduler.StepLR(
                    optimizer_det, step_size=5, gamma=0.5)

EPOCHS    = 20
best_loss = float('inf')
grid_sizes= {'large': 13, 'medium': 26, 'small': 52}

print("Starting training from scratch...")
print(f"Train samples: {len(train_dataset)}")
print(f"Epochs: {EPOCHS}\n")

for epoch in range(EPOCHS):
    model_det.train()
    total_loss = 0
    batch_count= 0

    for imgs, targets in train_loader:
        imgs    = imgs.to(device)
        targets = targets.to(device)

        optimizer_det.zero_grad()
        out_l, out_m, out_s = model_det(imgs)

        loss_l, *_ = compute_loss(out_l, targets,
                     ANCHORS['large'],  grid_sizes['large'])
        loss_m, *_ = compute_loss(out_m, targets,
                     ANCHORS['medium'], grid_sizes['medium'])
        loss_s, *_ = compute_loss(out_s, targets,
                     ANCHORS['small'],  grid_sizes['small'])

        loss = loss_l + loss_m + loss_s
        loss.backward()
        torch.nn.utils.clip_grad_norm_(model_det.parameters(), 1.0)
        optimizer_det.step()

        total_loss  += loss.item()
        batch_count += 1

        if batch_count % 50 == 0:
            print(f"  Epoch {epoch+1} | "
                  f"Batch {batch_count}/{len(train_loader)} | "
                  f"Loss: {loss.item():.4f}", end='\r')

    avg_loss = total_loss / batch_count
    scheduler_det.step()

    print(f"Epoch {epoch+1:02d}/{EPOCHS} | "
          f"Avg Loss: {avg_loss:.4f}        ")

    if avg_loss < best_loss:
        best_loss = avg_loss
        torch.save(model_det.state_dict(),
                   '/kaggle/working/vehicle_detector_best.pt')
        print(f"  ✅ Best model saved (loss: {avg_loss:.4f})")

print(f"\n🏆 Training complete. Best loss: {best_loss:.4f}")
```

```python
import torch
import numpy as np
from collections import defaultdict

model_det.load_state_dict(torch.load('/kaggle/working/vehicle_detector_best.pt'))
model_det.eval()

CONF_THRESH = 0.3
IOU_THRESH  = 0.5

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def decode_predictions(output, anchors, grid_size, img_size=416):
    """Decode raw model output into boxes"""
    B   = output.shape[0]
    A   = 3
    out_ch = 5 + NUM_CLASSES

    pred = output.view(B, A, out_ch, grid_size, grid_size)
    pred = pred.permute(0,1,3,4,2).cpu().numpy()

    boxes = []
    for b in range(B):
        img_boxes = []
        for a in range(A):
            aw, ah = anchors[a]
            for gy in range(grid_size):
                for gx in range(grid_size):
                    tx, ty, tw, th, obj = pred[b, a, gy, gx, :5]
                    cls_probs = pred[b, a, gy, gx, 5:]

                    obj_conf  = sigmoid(obj)
                    if obj_conf < CONF_THRESH:
                        continue

                    # Decode box
                    cx = (sigmoid(tx) + gx) / grid_size
                    cy = (sigmoid(ty) + gy) / grid_size
                    w  = (np.exp(tw) * aw) / img_size
                    h  = (np.exp(th) * ah) / img_size

                    cls_id   = np.argmax(cls_probs)
                    cls_conf = sigmoid(cls_probs[cls_id])
                    score    = obj_conf * cls_conf

                    if score > CONF_THRESH:
                        img_boxes.append([cls_id, score, cx, cy, w, h])
        boxes.append(img_boxes)
    return boxes

def box_iou(b1, b2):
    """b1, b2: [cx, cy, w, h]"""
    b1_x1 = b1[0] - b1[2]/2; b1_x2 = b1[0] + b1[2]/2
    b1_y1 = b1[1] - b1[3]/2; b1_y2 = b1[1] + b1[3]/2
    b2_x1 = b2[0] - b2[2]/2; b2_x2 = b2[0] + b2[2]/2
    b2_y1 = b2[1] - b2[3]/2; b2_y2 = b2[1] + b2[3]/2

    ix1 = max(b1_x1, b2_x1); ix2 = min(b1_x2, b2_x2)
    iy1 = max(b1_y1, b2_y1); iy2 = min(b1_y2, b2_y2)
    inter = max(0, ix2-ix1) * max(0, iy2-iy1)

    b1_area = (b1_x2-b1_x1) * (b1_y2-b1_y1)
    b2_area = (b2_x2-b2_x1) * (b2_y2-b2_y1)
    union   = b1_area + b2_area - inter + 1e-16
    return inter / union

# ── Compute per-class AP ──────────────────────────────────
all_detections = defaultdict(list)   # cls -> [(score, tp)]
all_gt_counts  = defaultdict(int)    # cls -> total gt boxes

anchor_list = {
    13: [(116,90),(156,198),(373,326)],
    26: [(30,61),(62,45),(59,119)],
    52: [(10,13),(16,30),(33,23)],
}

print("Computing mAP on validation set...")

with torch.no_grad():
    for batch_idx, (imgs, targets) in enumerate(val_loader):
        imgs = imgs.to(device)
        out_l, out_m, out_s = model_det(imgs)

        # Decode predictions from all scales
        preds_l = decode_predictions(out_l, anchor_list[13], 13)
        preds_m = decode_predictions(out_m, anchor_list[26], 26)
        preds_s = decode_predictions(out_s, anchor_list[52], 52)

        B = imgs.shape[0]
        for b in range(B):
            # Merge predictions from all scales
            all_preds = preds_l[b] + preds_m[b] + preds_s[b]

            # Get ground truth for this image
            gt_boxes = []
            for t in range(targets.shape[1]):
                if targets[b, t, 3] > 0:
                    cls = int(targets[b, t, 0].item())
                    # Only count classes we care about
                    if cls < NUM_CLASSES:
                        gt_boxes.append(targets[b, t].cpu().numpy())
                        all_gt_counts[cls] += 1

            # Match predictions to ground truth
            matched_gt = set()
            # Sort by confidence
            all_preds.sort(key=lambda x: -x[1])

            for pred in all_preds:
                cls_id, score, cx, cy, w, h = pred
                cls_id = int(cls_id)

                best_iou = 0
                best_gt  = -1
                for gi, gt in enumerate(gt_boxes):
                    if int(gt[0]) != cls_id:
                        continue
                    iou = box_iou([cx,cy,w,h],
                                  [gt[1],gt[2],gt[3],gt[4]])
                    if iou > best_iou:
                        best_iou = iou
                        best_gt  = gi

                if best_iou >= IOU_THRESH and best_gt not in matched_gt:
                    all_detections[cls_id].append((score, 1))  # TP
                    matched_gt.add(best_gt)
                else:
                    all_detections[cls_id].append((score, 0))  # FP

        if batch_idx % 2 == 0:
            print(f"  Batch {batch_idx+1}/{len(val_loader)}", end='\r')

# ── Compute AP per class ──────────────────────────────────
def compute_ap(detections, n_gt):
    if n_gt == 0 or len(detections) == 0:
        return 0.0
    detections.sort(key=lambda x: -x[0])
    tp_cum = 0
    fp_cum = 0
    precisions = []
    recalls    = []
    for score, tp in detections:
        if tp:
            tp_cum += 1
        else:
            fp_cum += 1
        precisions.append(tp_cum / (tp_cum + fp_cum))
        recalls.append(tp_cum / n_gt)

    # Area under PR curve
    ap = 0
    for i in range(1, len(precisions)):
        ap += (recalls[i] - recalls[i-1]) * precisions[i]
    return ap

print("\n" + "=" * 55)
print("     VEHICLE DETECTOR — mAP EVALUATION")
print("=" * 55)
print(f"  Confidence threshold : {CONF_THRESH}")
print(f"  IoU threshold        : {IOU_THRESH}")
print(f"  Epochs trained       : 10 / 20")
print("-" * 55)

aps = []
for cls_id, cls_name in enumerate(CLASS_NAMES):
    n_gt = all_gt_counts[cls_id]
    dets = all_detections[cls_id]
    ap   = compute_ap(dets, n_gt) if n_gt > 0 else 0.0
    aps.append(ap)
    print(f"  {cls_name:12s} | GT boxes: {n_gt:4d} | AP: {ap*100:.2f}%")

mAP = np.mean(aps)
print("-" * 55)
print(f"  mAP@0.5              : {mAP*100:.2f}%")
print(f"  Training status      : 10/20 epochs")
print(f"  Expected after 20ep  : ~{min(mAP*100*1.4, 85):.1f}%")
print("=" * 55)
```
